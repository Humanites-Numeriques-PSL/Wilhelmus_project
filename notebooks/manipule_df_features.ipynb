```python
import pandas as pd
```

```python
df_index = pd.read_csv("index_auteur.csv")
```

```python
dict_index = dict(zip(df_index['Texte'], df_index['Auteur']))
```

```python
df_main = pd.read_csv(r'features_train.csv')
```

```python
df_main.set_index("index", inplace = True)
```

```python
df_main
```

```python
list_index = list(df_main.index)
```

```python
list_index
```

```python
dict_index.pop("index_auteurs_txt-checkpoint")
```

```python
def convert_keys(dict_index):
    list_keys = []
    for keys in dict_index.keys():
        list_keys.append(int(keys))
    return list_keys
```

```python
list_keys = convert_keys(dict_index)
```

```python
dict_valide = dict(zip(list_keys, dict_index.values()))
```

```python
dict_valide
```

```python
def index_auteur(list_index, dict_valide):
    list_result = []
    for index in list_index:
        if index in dict_valide.keys():
            list_result.append(dict_valide[index])
        else:
            list_result.append('Anonymous')
    return list_result
```

```python
list_result = index_auteur(list_index, dict_valide)
```

```python
list_result
```

```python
df_main.index = list_result
```

```python
df_main
```

```python
df_main.to_csv(r'features_auteurs.csv', index = True)
```
